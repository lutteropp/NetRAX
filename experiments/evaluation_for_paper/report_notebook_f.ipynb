```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

Select the dataset name to evaluate:

```python
result_files = [
    "F_mpi_results.csv",
    "F_mpi_results2.csv"
]
#indices go from 0 to 1

filename = result_files[0]
filename2 = result_files[1]
```

```python
print("This evaluation is for: " + filename + " and " + filename2)
```

```python
df = pd.read_csv("result_files/"+filename)

df_average = df.query('likelihood_model == "AVERAGE"')
df_best = df.query('likelihood_model == "BEST"')

df_average.boxplot(column=['1', '2', '4', '8', '16', '32', '64'])
plt.suptitle("Runtime in seconds with LikelihoodModel.AVERAGE")
plt.xlabel("Number of MPI processes")
plt.tight_layout()
plt.savefig('F_mpi_runtime_average.png')
plt.show()
plt.suptitle("Runtime in seconds with LikelihoodModel.BEST")
df_best.boxplot(column=['1', '2', '4', '8', '16', '32', '64'])
plt.xlabel("Number of MPI processes")
plt.tight_layout()
plt.savefig('F_mpi_runtime_best.png')
plt.show()
```

```python
df = pd.read_csv("result_files/"+filename2)

ax = plt.gca()
df.plot(kind='line',x='n_procs',y='LikelihoodType.BEST',ax=ax,xticks=[16,32,48,64,80,96,112,128], style='.--', markersize=20)
df.plot(kind='line',x='n_procs',y='LikelihoodType.AVERAGE', color='red', ax=ax,xticks=[16,32,48,64,80,96,112,128], style='.--', markersize=20)

plt.suptitle("Runtime in seconds")
plt.xlabel("Number of MPI processes")
plt.tight_layout()
plt.savefig('F_mpi_runtime2.png')
plt.show()
```
