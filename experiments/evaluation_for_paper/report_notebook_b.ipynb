```python
%matplotlib inline
import matplotlib.pyplot as plt
from create_plots import import_dataframe, show_stats, show_plots

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

Select the dataset name to evaluate:

```python
result_files = [
    "B_20_1_results.csv",
    "C_20_1_results.csv",
    "D_scramble_results.csv",
    "E_psize_results.csv"
]
#indices go from 0 to 3

filename = result_files[0]
```

```python
print("This evaluation is for: " + filename)
```

```python
df = import_dataframe("result_files/"+filename)
filename = filename.split("_results.csv")[0]

def extract_prob(name):
    name = name.split("_")[-1]
    if name == "0":
        return 0.1
    elif name == "1":
        return 0.2
    elif name == "2":
        return 0.3
    elif name == "3":
        return 0.4
    elif name == "4":
        return 0.5
    return 42
    

df['reticulation_prob'] = df['name'].apply(lambda x: extract_prob(x))

df_prob_0_1 = df.query('reticulation_prob == 0.1')
df_prob_0_2 = df.query('reticulation_prob == 0.2')
df_prob_0_3 = df.query('reticulation_prob == 0.3')
df_prob_0_4 = df.query('reticulation_prob == 0.4')
df_prob_0_5 = df.query('reticulation_prob == 0.5')

print(df.columns)
```

# Original Simulated Dataset Statistics 

```python
show_stats(filename + "_prob_0_1", df_prob_0_1)
show_stats(filename + "_prob_0_2", df_prob_0_2)
show_stats(filename + "_prob_0_3", df_prob_0_3)
show_stats(filename + "_prob_0_4", df_prob_0_4)
show_stats(filename + "_prob_0_5", df_prob_0_5)
```
